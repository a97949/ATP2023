```python
import random

lista = []

def criar_lista():
    global lista
    tamanho = int(input("Quantos números tem a lista? "))
    lista = [random.randint(1, 100) for _ in range(tamanho)]
    print("Lista criada com sucesso!")

def ler_lista():
    global lista
    lista = []
    n = int(input("Quantos números quer inserir na lista? "))
    for i in range(n):
        numero = int(input(f"Insira o número {i + 1}: "))
        lista.append(numero)
    print("Lista lida com sucesso!")

def calcular_soma():
    global lista
    soma = sum(lista)
    print(f"Soma dos elementos da lista: {soma}")

def calcular_media():
    global lista
    media = sum(lista) / len(lista)
    print(f"Média dos elementos da lista: {media:.2f}")

def encontrar_maior():
    global lista
    maior = max(lista)
    print(f"Maior elemento da lista: {maior}")

def encontrar_menor():
    global lista
    menor = min(lista)
    print(f"Menor elemento da lista: {menor}")

def esta_ordenada_crescente():
    global lista
    if lista == sorted(lista):
        print("A lista está ordenada por ordem crescente: Sim")
    else:
        print("A lista está ordenada por ordem crescente: Não")

def esta_ordenada_decrescente():
    global lista
    if lista == sorted(lista, reverse=True):
        print("A lista está ordenada por ordem decrescente: Sim")
    else:
        print("A lista está ordenada por ordem decrescente: Não")

def procurar_elemento():
    global lista
    elemento = int(input("Que elemento quer procurar na lista? "))
    if elemento in lista:
        posicao = lista.index(elemento)
        print(f"Elemento {elemento} encontrado na posição {posicao}")
    else:
        print(f"(-1)")

def main():
    while True:
        print("\nMenu:")
        print("(1) Criar Lista")
        print("(2) Ler Lista")
        print("(3) Soma")
        print("(4) Média")
        print("(5) Maior")
        print("(6) Menor")
        print("(7) Esta Ordenada por Ordem Crescente")
        print("(8) Esta Ordenada por Ordem Decrescente")
        print("(9) Procurar um Elemento")
        print("(0) Sair")

        escolha = input("Escolha uma opção: ")

        if escolha == "1":
            criar_lista()
        elif escolha == "2":
            ler_lista()
        elif escolha == "3":
            calcular_soma()
        elif escolha == "4":
            calcular_media()
        elif escolha == "5":
            encontrar_maior()
        elif escolha == "6":
            encontrar_menor()
        elif escolha == "7":
            esta_ordenada_crescente()
        elif escolha == "8":
            esta_ordenada_decrescente()
        elif escolha == "9":
            procurar_elemento()
        elif escolha == "0":
            print("Lista atual:", lista)
            break
        else:
            print("Opção inválida. Escolha uma opção válida.")

if __name__ == "__main__":
    main()
```
